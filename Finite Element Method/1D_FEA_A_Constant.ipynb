```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def A(r):
  return (r+1)**2
```

```python
# Constants
Ti = 15 # K
To = 5 # K
k = 1.4 # W/mK
L = 5*10**(-3) # m 
B = 1 # m^2
node = 10
intervall = node - 1
delta_x = L/(intervall)
r = 0

print(delta_x)
```

```python
zero_data = np.zeros(shape=(node,node))
st_mat = pd.DataFrame(zero_data, columns= [i for i in range(1,node + 1)], index = [i for i in range(1,node + 1)])
st_mat
```

```python
for i in range(1,node):
  matrix = pd.DataFrame([[k*A(r)/delta_x, -k*A(r)/delta_x], [-k*A(r)/delta_x, k*A(r)/delta_x]], index = [i,i+1], columns = [i,i+1])
  st_mat = matrix.add(st_mat, fill_value=0)
  r += delta_x
st_mat 
```

```python
q_vec = np.zeros((node,1))
q_vec
```

```python
q_vec[1] = - (st_mat.iloc[1,0]  * Ti)
q_vec
```

```python
q_vec[-2] = (st_mat.iloc[3,2] * To)
q_vec
```

```python
st_mat = st_mat.iloc[1:-1,1:-1]
```

```python
q_vec = q_vec[1:-1]
```

```python
q_vec
```

```python
np.linalg.solve(st_mat, q_vec)
```
