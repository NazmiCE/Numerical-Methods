Numeric Solution

```python

import matplotlib.pyplot as plt
v = float(input("Velocity: "))
t0 = 0
tf = 10
dt = 2
fric = 12.5
m = 68.1
g = 9.8
velocity_list = [v]
time_list = [t0]
interval = (tf - t0)/dt
for i in range(int(interval)):
  v = v + ((m*g - v*fric)/m)*dt
  # print(v)
  velocity_list.append(v)
  time_list.append(time_list[-1] + dt)

plt.plot(time_list, velocity_list)
```

Converges to 53.39

Analitic Solution

```python
from math import exp
v = float(input("Velocity: "))
t0 = 0
tf = 10
dt = 0.01
fric = 12.5
m = 68.1
g = 9.8
velocity_list_a = [v]
time_list_a = [t0]
interval = (tf - t0)/dt
for i in range(int(interval)):
  v = (g*m)/fric*(1-exp((-1)*fric*(t0 + dt*i)/m))
  velocity_list_a.append(v)
  time_list_a.append(time_list_a[-1] + dt)
  # print(v)
plt.plot(time_list_a, velocity_list_a)
```
